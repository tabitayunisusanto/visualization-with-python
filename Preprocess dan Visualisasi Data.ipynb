### Terdapat Missing Value pada 7 variabel sehingga dilakukan Imputation

```python
data['ApplicantIncome'] = data['ApplicantIncome'].fillna((data['ApplicantIncome'].median())) 
data['CoapplicantIncome'] = data['CoapplicantIncome'].fillna((data['CoapplicantIncome'].median())) 
data['LoanAmount'] = data['LoanAmount'].fillna((data['LoanAmount'].mean())) 
data['Loan_Amount_Term'] = data['Loan_Amount_Term'].fillna((data['Loan_Amount_Term'].mean())) 
np.sum(data.isnull())
```

```python
data['Gender']=data['Gender'].fillna(data['Gender'].mode()[0])
data['Married']=data['Married'].fillna(data['Married'].mode()[0])
data['Dependents']=data['Dependents'].fillna(data['Dependents'].mode()[0])
data['Self_Employed']=data['Self_Employed'].fillna(data['Self_Employed'].mode()[0])
data['Credit_History']=data['Credit_History'].fillna(data['Credit_History'].mode()[0])
np.sum(data.isnull())
```

```python
data.describe(include='all')
```

### Encode Labels

```python
from sklearn.preprocessing import LabelEncoder
data['Credit_History'] = LabelEncoder().fit_transform(data['Credit_History'])
data.head(n=10)
```

### *Barchart* diatas menunjukkan Jumlah dari masing masing kategori pada variabel yang bertipe data kategorikal yaitu *Gender*, *Credit History*, *Self Employed*, *Education*, *Dependents*, dan Status Pernikahan

### Identifikasi Outlier dengan Standar Deviasi

### Jumlah Outlier pada Applicant Income:

```python
from numpy.random import seed
from numpy.random import randn
from numpy import mean
from numpy import std
data1=data.ApplicantIncome
data_mean, data_std = mean(data1), std(data1)
cut_off = data_std * 3
lower, upper = data_mean - cut_off, data_mean + cut_off
# identify outliers
outliers = [x for x in data1 if x < lower or x > upper]
print('Identified outliers: %d' % len(outliers))
# remove outliers
outliers_removed = [x for x in data1 if x >= lower and x <= upper]
print('Non-outlier observations: %d' % len(outliers_removed))
```
